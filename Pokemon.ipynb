```python
#extract the file from the downloaded zip file
zipfile_name = 'pokemon.zip'
with zipfile.ZipFile(zipfile_name, 'r') as file:
    file.extractall()
```

```python
#read in the csv file as pandas dataframe
pokemon = pd.read_csv("pokemon.csv")
```

```python
#Checking for missing values, not all pokemon have a second type
pokemon.isnull().sum()
```

```python
#Create data dictionary for mapping 1-6 to the actual name of region 
Generation_dict = {
    '1':'Kanto',
    '2':'Johto',
    '3':'Hoenn',
    '4':'Sinnoh',
    '5':'Unova',
    '6':'Kalos'
}
#Change data type of Generation column to string
pokemon.Generation = pokemon.Generation.astype('str')
#Map the values 1-6 to the name of the corresponding Region
pokemon.Generation = pokemon.Generation.map(Generation_dict)
```

```python
#Check if data mapping worked in dataframe
pokemon.head()
```

```python
#writing the final dataframe to an excel file for our tableau visualizations
pokemon.to_excel('pokemon_final.xlsx', sheet_name='Data')
```
