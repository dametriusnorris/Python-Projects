```python
#import packages
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'/Users/dametriusnorris/Documents/Projects/movies.csv')
```

```python
#Now we will look at the data
#use df.head() to show preview of first 5 rows of data 
df.head()
```

```python
#Locate missing data
##loop through the data and see if there is missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
```

```python
# Listing out the data types for each columns

print(df.dtypes)
```

```python
df.drop_duplicates()
```

```python
# Order our data by highest grossing film to lowest grossing film

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#View a correlation between gross and budget
sns.regplot(x="gross", y="budget", data=df)
```

```python
#See correlation between score and gross

sns.regplot(x="score", y="gross", data=df)
```

```python
# Correlation Matrix between all numeric columns to summarize data

df.corr(method = 'pearson')
```

```python
df.corr(method = 'kendall')
```

```python
df.corr(method = 'spearman')
```

```python
correlation_matrix = df.corr(method = 'spearman')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation Matrix for Numeric Features")

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```

```python
#Assign a random numeric value to each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation Matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
